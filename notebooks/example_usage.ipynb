# Image Captioning Example

This notebook demonstrates how to use the image captioning model locally.

```python
from PIL import Image
# Ensure you have installed the package or added src/ to path
import sys
import os
sys.path.insert(0, '../src')

from image_caption.models import generate_caption
import requests
from io import BytesIO
```

```python
# Load an image
url = "https://buffer.com/library/content/images/size/w1200/2023/10/free-images.jpg"
response = requests.get(url)
image = Image.open(BytesIO(response.content))

# Display image (if running in Jupyter)
image.thumbnail((400, 400))
display(image) if 'display' in locals() else print("Image loaded")
```

```python
try:
    caption = generate_caption(image)
    print(f"Caption: {caption}")
except Exception as e:
    print(f"Error: {e}")
```
